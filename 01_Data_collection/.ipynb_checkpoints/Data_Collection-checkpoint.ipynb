## Pull Data with REST API

```python
# import package for REST API
import requests
```

```python
url = "https://files.consumerfinance.gov/ccdb/complaints.json"
# url = "https://www.consumerfinance.gov/data-research/consumer-complaints/search/api/v1/?date_received_max=2022-07-25&date_received_min=2011-12-01&field=all&format=json&issue=Problems%20at%20the%20end%20of%20the%20loan%20or%20lease%E2%80%A2Problem%20related%20to%20refinancing&lens=overview&no_aggs=true&product=Vehicle%20loan%20or%20lease&size=467&trend_depth=5&trend_interval=month"

r = requests.get(url)
r.status_code
result_car_loan = r.json()
```

```python
result_car_loan
```

```python
# Check type of data
print(type(result_car_loan))
```

```python
result_car_loan_source = []
for i in result_car_loan:
    result_car_loan_source.append(i["_source"])
```

```python
result_car_loan_source
```

## convert to Pandas

```python
import pandas as pd
result_car_loan_source = pd.DataFrame(result_car_loan_source)
```
