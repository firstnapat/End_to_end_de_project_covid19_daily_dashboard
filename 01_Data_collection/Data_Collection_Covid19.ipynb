# Data Collection with REST API

## Requests library

```python
# import package for REST API
import requests
```

### pull data case all

```python
# requests data 
base_url = "https://covid19.ddc.moph.go.th/api/Cases/"
search_url = "timeline-cases-all"
response_cases_all = requests.get(base_url+search_url)
response_cases_all.status_code
covid_cases_all = response_cases_all.json()
```

```python
covid_cases_all
```

```python
# Check type of data
print(type(covid_cases_all))
```

## convert to pandas case all

```python
import pandas as pd
covid_cases_all = pd.DataFrame(covid_cases_all)
```

## Data profiling case all

```python
covid_cases_all.head()
```

```python
# ดูว่ามีคอลัมน์อะไรบ้าง
covid_cases_all.columns
```

#### Columns:
1. txn_date = transaction date
2. new_case = จำนวนผู้ติดเชื้อรายใหม่
3. total_case = จำนวนผู้ติดเชื้อทั้งหมด
4. new_case_excludeabroad = จำนวนผู้ติดเชื้อรายใหม่ไม่รวมต่างประเทศ
5. total_case_excludeabroad = จำนวนผู้ติดเชื้อทั้งหมดไม่รวมต่างประเทศ
6. new_death = จำนวนผู้เสียชีวิตรายใหม่
7. total_death = จำนวนผู้เสียชีวิตทั้งหมด
8. new_recovered = จำนวนผู้ป่วยที่รักษาหายรายใหม่
9. total_recovered = จำนวนผู้ป่วยที่รักษาหายทั้งหมด
10. update_date = วันที่อัพเดทข้อมูล
